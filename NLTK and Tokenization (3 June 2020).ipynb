# NLTK, Tokenization and introduction to Corpus (NLTK database)

```python
#importing nltk library
#NLTK is a NLP library which is used for text classification and text mining.
import nltk
```

```python
# #downloaded all the NLTK packages. This step is done once only.
# nltk.download()
```

```python
# assinging a text to a varible
# NOTE: Nltk will only recognize a new sentence using new line character or even regular expresions! and to do that we use triple quotes!
Python = """Python is the hight level, platform independent and open source. It is easy to understand.
It has mostly english keyword."""
```

```python
# importing word tokenizer to break the text into words as tokens
from nltk.tokenize import word_tokenize
```

```python
# saving the tokens into a variable
word = word_tokenize(Python)
```

```python
# list of all the tokens that is "words"
word
```

```python
# importing sentence tokenizer from nltk_tokenize package to break the text into sentences. This is done on the basis of new 
# line character
from nltk.tokenize import sent_tokenize
```

```python
# saving the values of tokens in variable
sent = sent_tokenize(Python)
```

```python
#returning a list of tokens that is "sentences"
sent
```

```python
# storing the values in "dream" variable
dream = nltk.corpus.shakespeare.words("dream.xml")
```

```python
# returning a list of words in the "shakespeare" file.
dream 
```
